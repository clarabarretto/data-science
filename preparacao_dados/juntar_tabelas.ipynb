```python
import pandas as pd


df1 = pd.read_csv('equipamentos_new.csv')
df2 = pd.read_csv('acidentes_2020-novo.csv', sep=';')
```

```python
df2.head(3)
```

```python
# inner join nas tabelas e salva o resultado
merged_df1_df2 = pd.merge(df1, df2, left_on='local_instalacao', right_on='endereco', how='inner')

merged_df1_df2.to_csv('equipamentos_acidentes.csv', index=False)
```
