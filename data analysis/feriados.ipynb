```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
diretorio = 'feriados-dataset/'

dfs = []

# Iterar sobre todos os arquivos no diretório
for filename in os.listdir(diretorio):
    if filename.endswith('.csv'):
        filepath = os.path.join(diretorio, filename)
        
        # Ler o arquivo CSV
        df = pd.read_csv(filepath, encoding='ISO-8859-1', sep=';')
        
        # Verificar se a coluna 'feriados' existe no DataFrame
        if 'feriados' in df.columns:
            # Dividir a coluna 'feriados' apenas na primeira ocorrência de ' - '
            df[['data', 'nome_feriado']] = df['feriados'].str.split(' - ', n=1, expand=True)
            
            # Remover a coluna 'feriados'
            df.drop(columns=['feriados'], inplace=True)
        
        # Filtrar os dados para 'Recife' e formatar a coluna 'data'
        if 'Cidade' in df.columns:
            filtered_df = df[df['Cidade'] == 'Recife']

            filtered_df['data'] = pd.to_datetime(filtered_df['data'], format='%d/%m/%Y').dt.strftime('%Y-%m-%d')

            dfs.append(filtered_df)
        else:
            print(f"Coluna 'Cidade' não encontrada no DataFrame {filename}")
    
# Concatenar todos os DataFrames da lista dfs em um único DataFrame final
df_final = pd.concat(dfs, ignore_index=True)

# Exibir o DataFrame final
df_final
```

```python
carnaval_1 = ['Recife', 'PE', '03-03-2025', 'Carnaval']
carnaval_2 = ['Recife', 'PE', '04-03-2025', 'Carnaval']

# Adicionando a nova linha usando loc[]
df_final.loc[len(df_final)] = carnaval_1
df_final.loc[len(df_final)] = carnaval_2

df_final
```

```python
df_final.to_csv('feriados-recife.csv')
```
